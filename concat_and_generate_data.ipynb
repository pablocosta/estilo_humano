```python
import pandas as pd


df_cino = pd.read_csv('./data/ustance/openai/cino.csv', sep=";")
df_cl = pd.read_fwf('./data/ustance/openai/cl.txt', header=None)
df_igreja = pd.read_fwf('./data/ustance/openai/igreja.txt', header=None)
```

```python
dfs = []
for path in ["./data/ustance/cino_test.csv", "./data/ustance/cino_train.csv"]:
    dfs.append(pd.read_csv(path, sep=";"))
df_real_cino = pd.concat(dfs, ignore_index=True)
df_real_cino["generated_Text"] = df_cino["Text"].tolist()



dfs = []
for path in ["./data/ustance/cl_test.csv", "./data/ustance/cl_train.csv"]:
    dfs.append(pd.read_csv(path, sep=";"))
df_real_cl = pd.concat(dfs, ignore_index=True)
df_real_cl["generated_Text"] = df_cl[0].tolist()

dfs = []
for path in ["./data/ustance/igreja_test.csv", "./data/ustance/igreja_train.csv"]:
    dfs.append(pd.read_csv(path, sep=";"))
df_real_igreja = pd.concat(dfs, ignore_index=True)
df_real_igreja["generated_Text"] = df_igreja[0].tolist()
```

```python
df_real_cino.to_csv("./data/ustance/final_cino.csv", sep="\t", index=False)
df_real_cl.to_csv("./data/ustance/final_cl.csv", sep="\t", index=False)
df_real_igreja.to_csv("./data/ustance/final_igreja.csv", sep="\t", index=False)
```
